## Obtención del diámetro de la columna.

Para obtener el valor del diámetro, se considerará la siguiente fórmula y las propiedades para el gas previamente mencionadas, se evalua la siguiente ecuación, para obtener un valor para el Número de Reynolds que cumpla con: $2100 \leq Re \leq 35000$. (Li et al., 2012) 

$$Re_g = \frac{\rho_g \cdot u \cdot D}{\mu_g}$$

De la cual se considera que:

$$u = \frac{\dot{V}}{A_{Transversal}}$$

Donde:

$$A_{Transversal} = \frac{\pi \cdot D^2}{4}$$

Sustituyendo las consideraciones tendremos lo siguiente:

$$Re_g = \frac{\rho_g \cdot \dot{V} \cdot D}{\mu_g \cdot A_{Transversal}}$$

$$Re_g = \frac{\rho_g \cdot \dot{V} \cdot D}{\mu_g \cdot \frac{\pi \cdot D^2}{4}}$$

La simplificación total de la ecuación tiene la siguiente forma:

$$Re_g = \frac{\rho_g \cdot \dot{V} \cdot 4}{\mu_g \cdot \pi \cdot D}$$

Con la ecuación obtenida, se grafica para evaluar y obtener el valor del número de Reynolds y el diámetro de la tubería. La evalución gráfica estará dada por el siguiente código:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

rho = 1.21 #kg/m3
mu = 1.75e-5 #Ns/m2
V = 0.037 #m3/s
pi = np.pi
D = np.arange(0.02, 0.82, 0.02)

def Re(D):
    return (rho * V * 4)/(mu * pi * D)

plt.figure(figsize=(12,5))
plt.plot(D, Re(D))
plt.xlabel("D [m]", fontsize=16)
plt.ylabel("Re", fontsize=16)
plt.title("Número de Reynols en función del diámetro de la columna.", fontsize=20)
plt.show()

TD = pd.DataFrame(Re(D), D, columns = ["Número de Reynolds"])
print("Los Resultados obtenidos por la evalución del número de Reynolds para cada posible diámetro está dado por: \n")
print(TD)

print("\nEl valor del diámetro para la columna de pared mojada a emplearse es de 0.4 m y el valor del número de Reynolds aproximado para el gas es: Re = " + str(round(Re(0.4),)) + ". Este flujo es turbulento y además cumple con estar dentro de los límites posteriormente definidos para el número de Reynolds.")
```

## Flujos.

Se toma el comportamiento del gas como un gas ideal. Con el que averiguo el flujo molar de entrada del $SO_2$:

$$P_{SO_2 In}\dot{V} = \dot{J}_{SO_2}^{In}RT$$

$$\dot{J}_{SO_2}^{In} = \frac{P_{SO_2 In}\dot{V}}{RT} = \frac{\left(4 \ kPa\right)\left(0.037 \ \frac{m^3}{s}\right)}{\left(8.314 \ \frac{kPa \cdot m^3}{kmol \cdot K}\right)\left(303.15 \ K\right)}$$

Como el enunciado nos indica que se debe eliminar el 90% del $SO_2$ presente en el aire. Podemos averiguar la cantidad de moles transferidos al agua, tomando el flujo molar de la entrada y multiplicandolo por el 90%, por lo que tendremos:

$$\dot{J}_{SO_2}^{Tranferidos} = \dot{J}_{SO_2}^{In} \cdot 90 \% = 5.28 \times 10^{-5} \ \frac{kmol}{s} = 5.28 \times 10^{-2} \ \frac{mol}{s}$$

Al conocer los moles transferidos al agua, se calcula el flujo molar en la salida de la columna:

$$\dot{J}_{SO_2}^{Out} = \dot{J}_{SO_2}^{In} - \dot{J}_{SO_2}^{Tranferidos} = 0.59 \times 10^{-5} \ \frac{kmol}{s}$$

Dado a que se asumió un flujo volumétrico constante a través de la columna. Aplicando la fórmula del gas ideal, se obtiende la presión parcial del $SO_2$ a la salida de la columna:

$$P_{SO_2 Out} = \frac{\dot{J}_{SO_2}^{Out}RT}{\dot{V}} = \frac{\left(0.59 \times 10^{-5} \ \frac{kmol}{s}\right)\left(8.314 \ \frac{kPa \cdot m^3}{kmol \cdot K}\right)\left(303.15 \ K\right)}{0.037 \ \frac{m^3}{s}}$$

## Fuerza motriz.

Con los valores obtenidos gracias al flujo de agua podemos calcular el valor para la fuerza motriz, la cual está dada por:

$$\Delta P_{LM} = \frac{\Delta P_1 - \Delta P_2}{ln\left(\frac{\Delta P_1}{\Delta P_2}\right)}$$

Donde los valores de para $\Delta P_1$ y $\Delta P_2$ están dados por:

$$\Delta P_1 = P_{SO_2 In} - P_{SO_2 Out}^{H_2O} = 4 \ kPa - 2.093 \ kPa = 1.907 \ kPa$$

$$\Delta P_2 = P_{SO_2 Out} - P_{SO_2 In}^{H_2O} = 0.40 \ kPa - 0 \ kPa = 0.40 \ kPa$$

Averiguamos el valor para $\Delta P_{LM}$:

$$\Delta P_{LM} = \frac{1.907 - 0.40}{ln\left(\frac{1.907}{0.40}\right)} = 0.965 \ kPa$$

Con esto, podemos definir la siguiente ecuación:

$$\dot{J}_{SO_2}^{Tranferidos}= K_GA\Delta P_{LM}$$

Despejando para hallar el valor de A se obtiene la siguiente ecuación:

$$A= \frac{\dot{J}_{SO_2}^{Tranferidos}}{K_G\Delta P_{LM}}$$

## Calculo de $K_G$

Para averiguar el valor de A, debemos encontrar el valor para $K_G$, por lo que tendremos lo siguiente:

### Coeficientes locales para el líquido.

Para fines de este trabajo se empleará un grosor de la película (L) de 0.145 mm, por ende el coeficiente de pleícula está dado por $\delta = 1.45 \times 10{-4} \ m$; (Karamian et al., 2019). Este está definido por la siguiente ecuación:

$$\delta = 1.45 \times 10{-4} \ m = \left(\frac{3\mu_w\Gamma}{\rho_w^2g}\right)^\frac{1}{3}$$

Despejando para hallar el valor de $\Gamma$, tenemos que:

$$\delta^3 = \frac{3\mu_w\Gamma}{\rho_w^2g}$$

$$\Gamma = \frac{\delta^3\rho_w^2g}{3\mu_w} = \frac{\left(1.45 \times 10{-4} \ m\right)^3\left(997 \ \frac{kg}{m^3}\right)^2\left(9.81 \ \frac{m}{s^2}\right)}{3\left(1.0 \times 10^{-3} \ \frac{N \cdot s}{m^2}\right)}$$

$$\Gamma = 0.009909 \ \frac{kg^2}{N \cdot s^3}$$

Con el valor obtenido para $\Gamma$ se averigua el número de Reynolds, que estará dado por la siguiente ecuación:

$$Re_L = \frac{4\Gamma}{\mu_w}$$

$$Re_L = \frac{4 \left(0.009909 \ \frac{kg^2}{N \cdot s^3}\right)}{1.0 \times 10^{-3} \ \frac{N \cdot s}{m^2}}$$

$$Re_L = 39.64 \approx 39$$

Dado a que el valor del número de Reynolds encontrado para el líquido es aproximadamente 89, por lo que se cumple lo siguiente: $Re < 100$. Por lo tanto, el valor para el número de Sherwood es: $Sh = 3.14$. (Green & Southard, 2019). Considerando la siguiente fórmula para el número de Sherwood, tenemos:

$$Sh = \frac{k_L\delta}{D_L} = 3.14$$

Despejando, obtenemos el valor para $k_L$:

$$k_L = \frac{ShD_L}{\delta}=\frac{3.14\left(1.7\times10^{-9} \ \frac{m^2}{s}\right)}{\left(1.45 \times 10{-4} \ m\right)}$$

$$k_L = 2.53\times10^{-5} \ \frac{m}{s}$$

### Coeficientes locales para el gas.

Sabemos que el valor para el número de Reynolds es: $Re_g = 8143$ (flujo turbulento). Asimismo, se calcula el valor para el número de Schmidt para el gas, que está dado por la siguiente ecuación:

$$Sc_g = \frac{\mu_g}{\rho_gD_g} = \frac{\left(1.75\times10^{-5} \ \frac{N\cdot s}{m^2}\right)}{\left(1.21 \ \frac{kg}{m^3}\right)\left(1.23\times10^{-5} \ \frac{m^2}{s}\right)}$$

$$Sc_g = 1.1758$$

Para el número de Sherwood, tendremos las siguientes expresiones, con las cuales, se despeja $k_g$.

$$Sh = k_g\frac{D}{D_g} = 0.023\left(Re_g\right)^{0.83}\left(Sc_g\right)^{0.44}$$

$$k_g = 0.023\left(Re_g\right)^{0.83}\left(Sc_g\right)^{0.44}\frac{D_g}{D}$$

$$k_g = 0.023 \cdot \left(8143\right)^{0.83} \cdot \left(1.1758\right)^{0.44} \cdot \frac{\left(1.23\times10^{-5} \ \frac{m^2}{s}\right)}{\left(0.4 \ m\right)}$$

$$k_g = 0.001338 \ \frac{m}{s}$$

## Cálculo de la altura de la columna.

Con la siguiente fórmula (propuesta anteriormente), al poseer A despejado, se sustituye para hallar el valor da A; Por lo que tendremos lo siguiente:

$$A = \frac{5.28\times10^{-2}}{\left(1.15 \times10^{-3}\right)\left(0.965\right)} = 47.57 \ m^2$$

La expresión para A, de la cual se despeja la altura (h), es la siguiente:

$$A = \pi D h$$

# Parte II.

Se  cuenta  con  una  columna  de  pared  húmeda  de  10  cm  de  diámetro  y  15  m  de  altura  ¿será capaz este equipo de lograr el objetivo planteado y eliminar el 90% del $SO_2$ aire? Para esto se realiza lo siguiente:

## Datos.

-  Diámetro (D): $10 \ cm = 0.1 \ m.$
-  Altura (h): $15 \ m.$
-  Debe cumplir con una extracción del 90% del $SO_2$ en el aire: $\dot{J}_{SO_2}^{Tranferidos} = 5.28 \times 10^{-2} \ \frac{mol}{s}$

## Resolución.

Para determinar si los parámetros de la columna de pared húmeda propuesta cumple con los requerimientos, se tendrá la siguiente fórmula:

$$\dot{J}_{SO_2}^{Tranferidos}= K_GA\Delta P_{LM}$$

Donde tendremos que:

-  Fuerza motriz ($\Delta P_{LM}$):

$$\Delta P_{LM} = 0.965 \ kPa$$

-  Coeficiente global de transferencia de masa ($K_G$):

$$K_G = 1.15 \times10^{-3} \ \frac{m}{s}$$

-  Área lateral de la columna (A):

$$A = \pi D h = \pi \left(0.1 \ m\right) \left(15 \ m\right) = 4.71 \ m^2$$

Calculamos el valor para los moles de $SO_2$ transferidos:

$$\dot{J}_{SO_2}^{Tranferidos}= \left(1.15 \times10^{-3} \ \frac{m}{s}\right)\left(4.71 \ m^2\right)\left(0.965 \ kPa\right)$$

$$\dot{J}_{SO_2}^{Tranferidos} = 5.23 \times10^{-3} \ \frac{mol}{s} \neq 5.28 \times10^{-2} \ \frac{mol}{s}$$

## Respuesta.

Dado a que el resultado obtenido para los moles transferidos de la columna de pared mojada propuesta por el enunciado no es igual a $5.28 \times10^{-2} \ \frac{mol}{s}$, se concluye que dicha columna no cumple con el parámetro solicitado de extracción de $SO_2$ del aire. Lo que implica que el equipo no sería capaz de eliminar el 90% del $SO_2$ aire.

# Información extra.

Acontinuación se presenta la gráfica que relaciona la variación del diámetro y altura requeridos para que se cumpla una eliminación del 90% del $SO_2$ aire, por lo que se empleará la siguiente fórmula para dicha graficación:

$$A= \pi D h = \frac{\dot{J}_{SO_2}^{Tranferidos}}{K_G\Delta P_{LM}}$$

Por ende la fórmula para el valor de la altura (h) en función del diámetro (D) es:

$$h(D) = \frac{\frac{\dot{J}_{SO_2}^{Tranferidos}}{K_G\Delta P_{LM}}}{\pi D}

El código para dicha graficación es el siguiente:


```python
D = np.arange(0.02, 0.82, 0.02)
KG = 1.15e-3 #m/s
PLM = 0.965 #kPa
J = 5.28e-2 #mol/s

def h(D):
    return ((J)/(KG*PLM))/(pi * D)

plt.figure(figsize=(12,5))
plt.plot(D, h(D))
plt.xlabel("D [m]", fontsize=16)
plt.ylabel("h [m]", fontsize=16)
plt.title("Altura de la columna en función del diámetro.", fontsize=20)
plt.show()

TH = pd.DataFrame(h(D), D, columns = ["Altura de la columna [m]"])
print("Los Resultados obtenidos por la evalución de la altura de la columna para cada posible diámetro está dado por: \n")
print(TH)
```
